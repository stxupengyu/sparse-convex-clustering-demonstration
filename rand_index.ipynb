```python
data=pd.read_csv('fit1.csv')
data=data.iloc[:,1:]
formater="{0:.01f}".format
data = data.applymap(formater)
data = data.apply(pd.to_numeric)
dic=data.drop_duplicates()
dic
```

```python
record=[]
i=0
while(i<144):
    for j in range(len(dic)):
        if(sum(data.iloc[i,:])==sum(dic.iloc[j,:])):
            record.append(j)
    i=i+1
real=[]
for i in range(6):
    temp=np.zeros(24)+i
    temp=temp.tolist()
    real=real+temp
def rand_index(y_true, y_pred):
    n = len(y_true)
    a, b = 0, 0
    for i in range(n):
        for j in range(i+1, n):
            if (y_true[i] == y_true[j]) & (y_pred[i] == y_pred[j]):
                a +=1
            elif (y_true[i] != y_true[j]) & (y_pred[i] != y_pred[j]):
                b +=1
            else:
                pass
    RI = (a + b) / (n*(n-1)/2)
    print(a,b,n)
    return RI
rand_index(real,record)
```
